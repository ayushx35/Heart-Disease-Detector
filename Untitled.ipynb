```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_data = pd.read_csv('/Users/macbookprofiles/Desktop/heart_disease/sample_data/heart_disease_data.csv')
```

```python
# 0-->healthy heart
# 1-->diseased heart
```

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_test,X_train,Y_test,Y_train=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_test.shape,X_train.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on Training Data = ",training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on Testing Data = ",test_data_accuracy)
```

```python
input_data=(62,0,0,140,268,0,0,160,0,3.6,0,2,2)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)
if(prediction[0]==0):
    print("person has a healthy heart")
else:
    print("person has a diseased heart")
```
